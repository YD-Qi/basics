# python的正确打开方式

我使用python，主要用于，加密货币行情的数据分析，策略开发以及机器人程序。

python是个大杀器，用处多多，此文记录我的一些使用经验（下面的操作都是基于mac的操作系统，pc系统会稍微有些不同）：

1. 下载最新版本 python

    官方网站： https://www.python.org/downloads/ 
    
    安装很简单，一直点击下一步直到安装完成。

2. 第三方库管理器

    我觉得最好用的是pip, 在mac terminal 运行如下两行代码即可：
    
    `curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py`

    `python3 get-pip.py`
    
    
3. 虚拟环境
    
    为每个项目新建虚拟环境，安装该项目所需第三方库，万一出现问题，可以删掉旧的虚拟环境，建一个新的，而不至于伤筋动骨整个python
    
    还是在mac terminal 窗口：
    
    * 这里我们用刚刚安装的第三方库管理器 `pip` 来安装虚拟环境的库：
    
    `pip install virtualenv`
    
    我们这里记住以后通过pip 安装第三方库都是这个命令格式 `pip install 第三方库名称`
    
    * 为新项目建一个文件夹：
    
    `mkdir my_project`
    
    `mkdir` 是新建路径的意思（make directory）, `my_project` 是新建文件夹的名字，任何名字都行。
    
    * 然后把新文件夹设定为当前路径，这样所有的操作默认都在新文件夹里：
    
    `cd ~user/my_project`
    
    `cd` 是改变路径（change directory）, user是自己电脑的用户名。
    
    * 这时我们就可以为新项目在其所在的文件夹设置虚拟环境了

     `virtualenv venv_tutorial`
     
      `virtualenv` 是命令， `venv_tutorial`是我们虚拟环境的名称，名称可以任意设置。
     
    * 下面这行代码启动虚拟环境
    
    `source venv_tutorial/bin/activate`
    
    这里注意需要根据前面你所设置的虚拟环境的名称做相应的修改，第一个斜线前，我这里是venv_tutorial


### 4. jupyter notebook


到这里，我们安装了最新的python, python第三方库管理器，以及如何为项目设置虚拟环境。

此时，我们距离正式写代码还有一个编译器（也叫 集成开发环境 Integrated development environment）的距离。

如果python是电脑主机的话，编译器就是显示器和键盘鼠标，让我们可以顺畅地使用这台电脑。

编译器选择有很多，pycharm, spyder, jupyter notebook。

我用的是jupyter notebook，我喜欢它的简介风格。结合它的markdown功能，可以将代码，输出结果和文字说明以漂亮的方式呈现出来。方便于学习笔记的整理和教程的编写。另外，jupyter notebook 也很适合数据分析，这对我做加密货币量化非常合适。这篇东西也是jupyter notebook写成的。

用jupyter notebook，可能大家有一个误区，就是离不开anaconda (https://www.anaconda.com/), 它是做数据科学的首选，通过它你可以打开r, python jupyter notebook 等等软件，功能强大。但是我感觉有时候并不好用，例如某个第三方库，如果不在anaconda的官方列表里，安装起来就会很麻烦。后来我找到了不依靠anaconda也可以使用jupyter notebook的方法，非常简单。

我们回到mac terminal窗口，运行以下代码：

`pip install jupyter`, 我们就安装好了jupyter notebook。

`ipython kernel install --user --name=venv_ytb`, 把虚拟环境加到jupyter notebook。

`jupyter notebook`, 随着浏览器窗口的打开，我们就进入到了jupyter notebook的页面，在这里我们可以新建一个文件，开启我们的python编程之路。
